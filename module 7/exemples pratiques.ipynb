### Requests

Le module requests permet d'envoyer simplement des demandes vers des pages web ou des API REST. Les API REST sont les successeurs des webservices et sont fréquement utilisées en raison de leur simplicité de mise en oeuvre.

```python
import requests as req
```

```python
# requête web simple

response = req.get("https://meritis.fr/blog/")
```

```python
# L'attibut status_code permet de savoir si une requête a aboutie
response.status_code
```

```python
# ce qui a été généré comme requête
response.request.__dict__
```

```python
# la réponse
response.text
```

```python
# pas très lisible, faisons mieux
import bs4
```

```python
soup = bs4.BeautifulSoup(response.text)
```

```python
soup
```

```python
print(soup.body.get_text())
```
