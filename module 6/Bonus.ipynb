```python
immo = pd.read_csv("https://cadastre.data.gouv.fr/data/etalab-dvf/latest/csv/2018/departements/78.csv.gz",
                   decimal='.')
immo = immo[immo.type_local.isin(['Maison','Appartement'])]
immo = immo.drop_duplicates(subset=['id_mutation'], keep=False)
immo = immo[ immo.valeur_fonciere < 4000000]
immo = immo[ immo.surface_reelle_bati < 500]
immo['prix_m2'] = immo.valeur_fonciere / immo.surface_reelle_bati
```

```python
immo.columns
```

```python
fig = ex.scatter_mapbox(immo[ immo.prix_m2 < 20000], lon='longitude', lat='latitude', color='prix_m2', hover_name='nom_commune')
fig.update_layout(mapbox_style="open-street-map")
```
